```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

breast_cancer = load_breast_cancer()
df1 = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df1['Target'] = breast_cancer.target
```

```python
data = load_breast_cancer()
X = data.data
Y = data.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,Y)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,Y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf,filled= True)
```

```python
print(classification_report(Y_test, clf.predict(X_test)))
```

```python
param = {
    'criterion': ['gini', 'entropy' ],
    'max_depth': [2, 3, 4, 5, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2'],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(
    estimator=clf,
    param_grid=param,
    scoring='accuracy',
    n_jobs=3)
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
best_models=grid.best_estimator_
```

```python
grid.best_score_
```

```python
best_models.fit(X_train,Y_train)
```

```python
print(classification_report(Y_test,best_models.predict(X_test)))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

ConfusionMatrixDisplay.from_estimator(best_models, X_test, Y_test)
plt.title("Confusion Matrix")
plt.show()
```
